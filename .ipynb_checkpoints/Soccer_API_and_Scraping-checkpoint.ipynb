## Adding coefficient and z-score

```python
from scipy import stats
```

```python
coef_df = df.copy(deep=True)
```

```python
coef_df.describe()
```

```python
coef_df.sort_values(by='total_points', ascending=False)
```

```python
coef_df['goal_z-score'] = stats.zscore(coef_df['goals_scored'])
```

```python
coef_df['goal_coef'] = coef_df['goal_z-score'] * 0.35
```

```python
coef_df['assist_z-score'] = stats.zscore(coef_df['assists'])
```

```python
coef_df['assist_coef'] = coef_df['assist_z-score'] * 0.20
```

```python
coef_df['value-form_z-score'] = stats.zscore(coef_df['value_form'])
```

```python
coef_df['value-form_coef'] = coef_df['value-form_z-score'] * 0.05
```

```python
coef_df['form_z-score'] = stats.zscore(coef_df['form'])
```

```python
coef_df['form_coef'] = stats.zscore(coef_df['form_z-score']) * 0.10
```

```python
coef_df['bps_z-score'] = stats.zscore(coef_df['bps'])
```

```python
coef_df['bps_coef'] = stats.zscore(coef_df['bps_z-score']) * 0.05
```

```python
coef_df['ict_z-score'] = stats.zscore(coef_df['ict_index'])
```

```python
coef_df['ict_coef'] = stats.zscore(coef_df['ict_z-score']) * 0.15
```

```python
coef_df.columns
```

```python
coef_df['total_coef'] = coef_df['goal_coef'] + coef_df['assist_coef'] + coef_df['value-form_coef'] +\
coef_df['coef_form'] + coef_df['bps_coef'] + coef_df['ict_coef']
```

```python
coef_df.sort_values(by='total_coef', ascending=False)
```

```python
coef_df[['web_name', 'ict_index', 'now_cost', 'total_coef', 'minutes',\
        'points_per_min']].sort_values(by='total_coef', ascending=False)
```

```python
#! We should find a way to calculate bang for buck:
# i.e., how many points players are getting per minute of playing time
```

```python
#! Current model screws defenders
# Perhaps we should include total points as one of the considerations?
# Would be neat if we could filter by player position, but the API does not include this, and there are
# more than 500 players
```

```python
coef_df['points_per_min'] = coef_df['total_points'] / coef_df['minutes']
```

```python
coef_df[coef_df['minutes'] > 90].sort_values(by='points_per_min', ascending=False)
```

```python
coef_df['ppm/now_cost'] = coef_df['points_per_min'] / coef_df['now_cost']
```

```python
coef_df[['web_name', 'ict_index', 'now_cost', 'total_coef', 'minutes',\
        'points_per_min', 'ppm/now_cost']][coef_df['minutes']>90].sort_values(by='ppm/now_cost', ascending=False)[0:40]
```

# Web Scrape from FoxSoccer

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://www.foxsports.com/soccer/stats?competition=1&season=20160&\
category=STANDARD&pos=None&team=0&isOpp=0&sort=3&sortOrder=0'
```

```python
r = requests.get(url)
r.content
```

```python
soup = BeautifulSoup(r.content, 'lxml')
print(soup.prettify())
```

```python
links = soup.find_all('a')
print(links)
```

```python
player_all_data = soup.find_all('tr', {'class': 'wisbb_fvStand'})# 'a', {'class': 'wisbb_fullPlayer'})
print(player_all_data)
```

```python
for player in player_all_data:
    print(player.contents[1].text)
    print(player.contents[9].text)
```

```python
for player in player_all_data:
    try:
        print(player.contents[1].find_all('span')[1].text)
    except:
        pass
```

```python
name = []
goals = []
assists = []

for player in player_all_data:
    try:
        name.append(player.contents[1].find_all('span')[1].text)
    except:
        pass
    goals.append(player.contents[9].text)
    assists.append(player.contents[11].text)

goals = goals[1:]
assists = assists[1:]
```

```python
players_2015_2016 = pd.DataFrame(
    {'Player Name': name,
    'Goals': goals,
    'Assists': assists})
```

```python
players_2015_2016.head()
```

```python
players_2015_2016 = players_2015_2016[['Player Name', 'Goals', 'Assists']]
```
